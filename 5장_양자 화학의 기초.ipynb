```python
import numpy as np
import matplotlib.pyplot as plt

# 사인 함수
def sine_wave(x, amplitude=1, frequency=1, phase=0):
    return amplitude * np.sin(2 * np.pi * frequency * x + phase)

# 코사인 함수
def cosine_wave(x, amplitude=1, frequency=1, phase=0):
    return amplitude * np.cos(2 * np.pi * frequency * x + phase)

# 가우시안 함수
def gaussian_wave(x, mean=0, standard_deviation=1):
    return (1 / (np.sqrt(2 * np.pi) * standard_deviation)) * np.exp(-0.5 * ((x - mean) / standard_deviation) ** 2)

# 사각파 함수
def square_wave(x, amplitude=1, frequency=1):
    return amplitude * np.sign(np.sin(2 * np.pi * frequency * x))

# 삼각파 함수
def triangle_wave(x, amplitude=1, frequency=1):
    return amplitude * np.abs(2 * (x * frequency - np.floor(x * frequency + 0.5))) - amplitude / 2

# x 값 범위
x_vals = np.linspace(-5, 5, 1000)

# 각 웨이브 함수에 대해 그래프 그리기
plt.plot(x_vals, sine_wave(x_vals), label='Sine Wave')
plt.plot(x_vals, cosine_wave(x_vals), label='Cosine Wave')
plt.plot(x_vals, gaussian_wave(x_vals), label='Gaussian Wave')
plt.plot(x_vals, square_wave(x_vals), label='Square Wave')
plt.plot(x_vals, triangle_wave(x_vals), label='Triangle Wave')

# 그래프 설정
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Different Wave Functions')
plt.grid(True)

# 그래프 표시
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 가우시안 함수
def gaussian_wave(x, mean=0, standard_deviation=1):
    return (1 / (np.sqrt(2 * np.pi) * standard_deviation)) * np.exp(-0.5 * ((x - mean) / standard_deviation) ** 2)

# 정규화 상수 계산 (수치 적분 사용)
def numerical_integration(func, a, b, num_points=1000):
    x_vals = np.linspace(a, b, num_points)
    dx = (b - a) / (num_points - 1)
    integral = np.sum([func(x) * dx for x in x_vals])
    return integral

# 정규화 상수 계산
normalization_constant = np.sqrt(numerical_integration(lambda x: abs(gaussian_wave(x))**2, -10, 10))

# 정규화된 가우시안 함수
def normalized_gaussian_wave(x):
    return gaussian_wave(x) / normalization_constant

# x 값 범위
x_vals = np.linspace(-5, 5, 1000)

# 정규화된 가우시안 웨이브 함수와 확률 밀도 그래프 그리기
plt.plot(x_vals, normalized_gaussian_wave(x_vals), label='Normalized Gaussian Wave Function')
plt.plot(x_vals, abs(normalized_gaussian_wave(x_vals))**2, label='Probability Density')

# 그래프 설정
plt.xlabel('x')
plt.ylabel('Amplitude / Probability Density')
plt.legend()
plt.title('Normalization and Probability Density of Gaussian Wave Function')
plt.grid(True)

# 그래프 표시
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 상자의 길이 (길이가 L인 상자에서 0 <= x <= L)
L = 1.0

# 파동 함수, 에너지, 확률 밀도를 계산하는 함수
def wave_function(n, x, L):
    # 파동 함수 (정규화된)
    return np.sqrt(2/L) * np.sin(n * np.pi * x / L)

def energy(n, L):
    # 에너지 준위 (단위: h_bar = m = 1)
    return (n**2 * np.pi**2) / (2 * L**2)

def probability_density(psi):
    # 확률 밀도
    return psi**2

# x 값의 범위
x_values = np.linspace(0, L, 400)

# 첫 3개의 양자 상태에 대해 파동 함수, 에너지, 확률 밀도를 그래프로 그리기
for n in range(1, 4):
    psi_values = wave_function(n, x_values, L)
    prob_density_values = probability_density(psi_values)

    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plt.plot(x_values, psi_values, label=f'n = {n}')
    plt.xlabel('x')
    plt.ylabel('Ψ(x)')
    plt.title(f'Wave Function for n = {n}')
    plt.grid(True)
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(x_values, prob_density_values, label=f'n = {n}')
    plt.xlabel('x')
    plt.ylabel('|Ψ(x)|^2')
    plt.title(f'Probability Density for n = {n}')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f'Energy for n = {n}: {energy(n, L)}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite
from math import pi, sqrt, exp, factorial

def harmonic_oscillator_wave_function(n, x):
    m = 1.0  # 질량
    omega = 1.0  # 각 주파수
    hbar = 1.0  # 감소된 플랑크 상수

    # 정규화 상수
    coeff = 1.0 / sqrt(2**n * factorial(n)) * (m * omega / (pi * hbar))**(1/4)
    # 허미트 다항식
    herm_poly = hermite(n)
    # 웨이브 함수 계산
    wave_function = coeff * herm_poly(np.sqrt(m * omega / hbar) * x) * np.exp(-m * omega * x**2 / (2 * hbar))

    return wave_function

x = np.linspace(-5, 5, 400)  # x 좌표 배열

plt.figure(figsize=(10, 5))

# n = 0, 1, 2, 3에 대한 웨이브 함수를 그립니다.
for n in range(4):
    psi = harmonic_oscillator_wave_function(n, x)
    plt.plot(x, psi, label=f'n={n}')

plt.xlabel('Position')
plt.ylabel('Wave Function')
plt.legend()
plt.title('Wave Functions of a Quantum Harmonic Oscillator')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import hbar
from scipy.special import eval_hermitenorm
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

def finite_well_wavefunction(x, n, a, V0, m):
    omega = np.sqrt(2 * V0 / (m * a**2))  # 조화 진동자의 각 주파수 계산
    normalization = (m * omega / (np.pi * hbar))**(1/4) / np.sqrt(2**n * np.math.factorial(n))
    hermite_val = eval_hermitenorm(n, np.sqrt(m * omega / hbar) * x)
    psi = normalization * hermite_val * np.exp(- (m * omega * x ** 2) / (2 * hbar))
    return psi

# 파라미터 설정
m = 9.109e-31  # 전자의 질량 (kg)
a = 1e-10  # 우물의 너비 (미터)
V0 = 10 * 1.6e-19  # 우물의 높이 (Joule)
alpha = 0.5  # 회색 영역의 투명도

# 위치 범위 설정
x_values = np.linspace(-5 * a, 5 * a, 1000)  # 위치 범위

# 에너지 준위 설정
n_values = list(range(0, 6))  # 에너지 준위 0부터 5까지

# 파동 함수 계산
psi_values = []
for n in n_values:
    psi = finite_well_wavefunction(x_values, n, a, V0, m)
    psi_values.append(psi)

# 파동 함수 시각화
fig, ax = plt.subplots(figsize=(12, 8))
for i, n in enumerate(n_values):
    ax.plot(x_values, np.abs(psi_values[i])**2, label=f'n = {n}', linewidth=1)

# 우물 영역 회색으로 표시
ax.fill_between([-a, a], 0, np.max(np.abs(psi_values))**2, color='gray', alpha=alpha)

# 부분 확대 그래프 추가
axins = inset_axes(ax, width="30%", height="30%", loc="center right")
for i, n in enumerate(n_values):
    axins.plot(x_values[x_values > 0], np.abs(psi_values[i][x_values > 0])**2, linewidth=0.7)
axins.fill_between([0, a], 0, np.max(np.abs(psi_values))**2, color='gray', alpha=alpha)

# 부분 확대 영역 설정
x1, x2, y1, y2 = 0, 1.7 * a, 0, np.max(np.abs(psi_values))**2 * 0.1
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)

plt.xlabel('Position (x)')
plt.ylabel('Probability Density')
plt.title('Finite Potential Well - Probability Density')
ax.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.special import sph_harm

# 구면조화함수
l = 2  # 각운동량양자수
m = 1  # 자기 양자수

# 각도 정의
theta = np.linspace(0, np.pi, 100)  # Polar angle
phi = np.linspace(0, 2*np.pi, 100)  # Azimuthal angle

# meshgrid생성
Theta, Phi = np.meshgrid(theta, phi)

# 구면조화함수 계산
Ylm = sph_harm(m, l, Phi, Theta)

# Catesian 좌표로 변환하여 플로팅 준비
X = np.abs(Ylm) * np.sin(Theta) * np.cos(Phi)
Y = np.abs(Ylm) * np.sin(Theta) * np.sin(Phi)
Z = np.abs(Ylm) * np.cos(Theta)

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, color='c', alpha=0.5, linewidth=0)

ax.set_title('Spherical Harmonics |Y^{}_{}|'.format(l, m))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm, genlaguerre
import math

def hydrogen_radial_wave_func(n, l, r):
    # 상수
    a0 = 1  # 보어 반지름

    # 정규화 상수
    norm_const = (2 / (n * a0)) ** 3 * (math.factorial(n - l - 1) / (2 * n * math.factorial(n + l))) ** 0.5

    # 연관 라게르 다항식
    rho = 2 * r / (n * a0)
    L = genlaguerre(n - l - 1, 2 * l + 1)(rho)

    # 파동 함수의 반지름 부분
    R = norm_const * np.exp(-rho / 2) * rho ** l * L

    return np.abs(R) ** 2

# r에 대한 격자 생성
r = np.linspace(0, 20, 1000)

# 양자 수 정의
quantum_numbers = [(1, 0, 0), (2, 0, 0), (2, 1, 0), (3, 0, 0), (3, 1, 0), (3, 2, 0)]

# 서브플롯 생성
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# 다양한 양자 수에 대한 반지름 파동 함수를 플롯
for n, l, m in quantum_numbers:
    ax1.plot(r, hydrogen_radial_wave_func(n, l, r), label=f'n={n}, l={l}, m={m}')
    ax2.plot(r, hydrogen_radial_wave_func(n, l, r), label=f'n={n}, l={l}, m={m}')

# 첫 번째 플롯 설정
ax1.set_title('수소 원자 반지름 파동 함수')
ax1.set_xlabel('r (반지름)')
ax1.set_ylabel('확률 밀도')
ax1.legend()
ax1.grid(True)

# 두 번째 플롯 설정 (y축 확대)
ax2.set_title('수소 원자 반지름 파동 함수 (확대)')
ax2.set_xlabel('r (반지름)')
ax2.set_ylabel('확률 밀도')
ax2.set_ylim(0, 0.05)  # y축 제한 설정 (확대된 플롯)
ax2.grid(True)

# 플롯 표시
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from scipy.special import sph_harm, genlaguerre
import math
from mpl_toolkits.mplot3d import Axes3D


def hydrogen_wave_func(n, l, m, r, theta, phi):
    # 상수
    a0 = 1  # 보어 반지름

    # 정규화 상수
    norm_const = (2 / (n * a0)) ** 3 * (math.factorial(n - l - 1) / (2 * n * math.factorial(n + l))) ** 0.5

    # 연관 라게르 다항식
    rho = 2 * r / (n * a0)
    L = genlaguerre(n - l - 1, 2 * l + 1)(rho)

    # 파동 함수의 반径 부분
    R = norm_const * np.exp(-rho / 2) * rho ** l * L

    # 파동 함수의 각도 부분
    Y = sph_harm(m, l, phi, theta)

    # 전체 파동 함수
    psi = R * Y

    return np.abs(psi) ** 2


# 양자 수 정의
quantum_numbers = [(2, 1, 0)]

# r, theta, phi에 대한 격자 생성
r = np.linspace(0, 20, 100)
theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2 * np.pi, 100)
r, theta, phi = np.meshgrid(r, theta, phi)

# 데카르트 좌표계로 변환
x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)

# 서로 다른 양자 수에 대한 파동 함수 그리기
fig = plt.figure(figsize=(8, 8))
for idx, (n, l, m) in enumerate(quantum_numbers, start=1):

    # 파동 함수 얻기
    psi_squared = hydrogen_wave_func(n, l, m, r, theta, phi)

    # 시각화를 위한 무작위 샘플링
    mask = np.random.random(psi_squared.shape) < 0.05

    # 플롯
    ax = fig.add_subplot(1, 1, idx, projection='3d')
    sc = ax.scatter(x[mask], y[mask], z[mask], c=psi_squared[mask], cmap=cm.plasma, alpha=0.01)

    ax.set_title(f'n={n}, l={l}, m={m}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    fig.colorbar(sc, ax=ax)

# 컬러 범위 조정
vmin, vmax = np.min(psi_squared), 0.000005
for ax in fig.axes:
    ax.collections[0].set_clim(vmin, vmax)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# STO 함수
def sto(r, Z=1, N=1, x=1, y=1, z=1, R=0):
    return N * x * y * z * np.exp(-Z * np.abs(r - R))
# GTO 함수
def gto(r, alpha=1, N=1, x=1, y=1, z=1, R=0):
    return N * x * y * z * np.exp(-alpha * (r - R)**2)
# r 값 범위 (예: -5 ~ 5)
r = np.linspace(-5, 5, 400)
# STO와 GTO 함수 값 계산
sto_values = sto(r)
gto_values = gto(r)
# 그래프 그리기
plt.plot(r, sto_values, label='STO')
plt.plot(r, gto_values, label='GTO')
plt.xlabel('r')
plt.ylabel('Function Value')
plt.title('STO and GTO functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from pyscf import gto, scf, cc

# Define the molecule
mol = gto.Mole()
mol.atom = "O 0 0 0; H 0 -0.757 0.587; H 0 0.757 0.587"
mol.basis = "sto-3g"
mol.build()

# Hartree-Fock 계산
mf = scf.RHF(mol)
mf.kernel()

# Coupled Cluster 계산
mycc = cc.CCSD(mf)
mycc.kernel()

# CCSD 에너지
energy_ccsd = mf.e_tot + mycc.e_corr
print(f"CCSD Energy: {energy_ccsd} Hartree")
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from pyscf import gto, scf

# RDKit을 사용하여 SMILES를 분자로 변환
smiles = 'CCO'  # 에탄올 (ethanol)
mol_rdkit = Chem.MolFromSmiles(smiles)
mol_rdkit = Chem.AddHs(mol_rdkit)  # 분자에 수소 원자를 추가
AllChem.EmbedMolecule(mol_rdkit)  # 분자의 3D 좌표 생성
AllChem.UFFOptimizeMolecule(mol_rdkit)  # UFF로 분자 최적화

# RDKit 분자를 PySCF 분자로 변환
atom_charges = [atom.GetAtomicNum() for atom in mol_rdkit.GetAtoms()]  # 원자 번호 가져오기
atom_coords = mol_rdkit.GetConformer().GetPositions()  # 원자 좌표 가져오기
atom_str = ""
for charge, coords in zip(atom_charges, atom_coords):
    atom_str += f"{charge} {coords[0]} {coords[1]} {coords[2]}; "

# PySCF 분자 생성
mol_pyscf = gto.Mole()
mol_pyscf.atom = atom_str
mol_pyscf.basis = 'sto-3g'  # 기본 함수 설정 (STO-3G)
mol_pyscf.build()

# Hartree-Fock 계산 수행
mf = scf.RHF(mol_pyscf)  # 제한된 Hartree-Fock 방법
energy = mf.kernel()

# 총 에너지 출력 (Hartree 단위)
print(f"총 에너지: {energy} Hartree")
```
